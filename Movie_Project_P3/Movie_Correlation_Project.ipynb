```python
# import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# read in data

df = pd.read_csv('movies.csv')
```

```python
# looking at the data

df.head()
```

## Data Cleaning

```python
# checking for missing data
# iterating through dataset to see what data is missing

for col in df.columns:
    missing = df[col].isnull().sum() 
    print('{} - {}'.format(col, round(missing)))
```

```python
# dropping the rows with null values
dfc = df.copy(deep= True)
dfc = df.dropna()

# checking to see if the null values have been successfully removed
for col in dfc.columns:
    missing = dfc[col].isnull().sum() 
    print('{} - {}'.format(col, round(missing)))
```

```python
# datatypes for the columns

dfc.dtypes
```

```python
# inconsistency between release year in 'year' vs 'released' column
# creating a corrected year column using the year values in the 'released' column since the values in 'year' did not match
dfc['yearcorrect'] = dfc['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# sorting the data from highest to lowest gross profit

dfc.sort_values(by=['gross'], inplace=False, ascending=False)
```

## Data Analysis

### Scatterplot

```python
# scatterplot - budget vs gross revenue
    
plt.scatter(x= dfc['budget'], y= dfc['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Film Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

### Regression Plot

```python
# regression plot using seaborn - budget vs gross

sns.regplot(x= 'budget', y= 'gross', data= dfc, scatter_kws={'color': 'red'}, 
            line_kws={'color': 'blue'})
```

### Correlation Matrix

```python
# correlation matrix

dfc.corr()
correlation_matrix = dfc.corr()

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

### Encoding 

```python
# encoding categorical values as numeric values
dfc_encoded = dfc.copy(deep=True)
for col_name in dfc_encoded.columns:
    if(dfc_encoded[col_name].dtype == 'object'):
        dfc_encoded[col_name] = dfc_encoded[col_name].astype('category')
        dfc_encoded[col_name] = dfc_encoded[col_name].cat.codes
      
dfc_encoded
```

```python
dfc
```

### Correlation Matrix for Encoded Features

```python
# correlation of encoded features
correlation_matrix = dfc_encoded.corr()

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Encoded Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
# correlations in text
corr_mtx = dfc_encoded.corr()
corr_pairs = corr_mtx.unstack()
corr_pairs
```

```python
# correlations sorted
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# showing the pairs with the highest correlations

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```

```python
# showing the pairs with the lowest correlations

low_corr = sorted_pairs[(sorted_pairs) < 0.5]
low_corr
```

## Findings

Gross profit and budget have the highlest correlation (0.615), while genre and budget have the lowest correlation (-0.369).

If a movie company wanted to make data-driven decisions regarding what makes a movie successful, they could consider that a higher budget may lead to the movie making a higher profit. However, they should also keep in mind that correlation doesn't necessarily imply causation.
